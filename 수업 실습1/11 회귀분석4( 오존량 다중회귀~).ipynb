### 오존량 예측 (다중회귀분석)

```python
import pandas as pd
df = pd.read_csv("C:/vscode/data/ozone.csv")
df.head()
```

```python
#오존량, 일조량 결측값 처리
%matplotlib inline
import missingno as msno
import matplotlib.pyplot as plt

msno.matrix(df)
#흰색 - 결측값
#스파크라인(spark line) - 각 샘플의 데이터 완성도를 표현
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="mean")
df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
df
```

```python
X=df[['Solar.R','Wind','Temp']]
y=df['Ozone']

import statsmodels.api as sm

model = sm.OLS(y, X)
result = model.fit()
result.summary()
```

```python
#일조량 92, 풍량 15.5, 화씨온도 84 일때의 오존량 예측
result.predict([92,15.5,84])
```

```python
# 예측모델 저장
result.save("C:/vscode/data/ozone/ozone_regress.model")
```

```python
#불러오기
from statsmodels.regression.linear_model import OLSResults
model = OLSResults.load("C:/vscode/data/ozone/ozone_regress.model")
```

```python
#일조량 92, 풍량 15.5, 화씨온도 84 일때의 오존량 예측
model.predict([92,15.5,84])
```
