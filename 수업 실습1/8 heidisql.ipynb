```python
# 항공사별 출발 지연 시간 평균 구하기
conn = MySQLdb.connect(host='localhost', port=3306, db='ontime',user='web', passwd='1234')

cursor = conn.cursor()                      
cursor.execute('''select uniquecarrier, avg(depdelay)
from ontime group by uniquecarrier''')    
rows = cursor.fetchall()          
for row in rows:
    print('{}\t{}'.format(row[0],row[1]))
    cursor.close()
conn.close()
```

```python
#2007년 도착지연횟수 월별 집계
conn = MySQLdb.connect(host='localhost', port=3306, db='ontime',user='web', passwd='1234')

cursor = conn.cursor()                      
cursor.execute('''select month, count(*) from ontime
where year=2007 and arrdelay > 0
group by month''')    
rows = cursor.fetchall()         
for row in rows:
    print(row)
    
cursor.close()
conn.close()
```

```python
#요일별 평균 지연 시간
conn = MySQLdb.connect(host='localhost', port=3306, db='ontime',user='web', passwd='1234')

cursor = conn.cursor()                      
cursor.execute('''select dayofweek, avg(depdelay)
from ontime
group by dayofweek''')    
rows = cursor.fetchall()         
for row in rows:
    print(row)

cursor.close()
conn.close()
```

```python
#시간대별( 10월, 월요일에 출발할 때 가장 좋은 시간대는? )
conn = MySQLdb.connect(host='localhost', port=3306, db='ontime',user='web', passwd='1234')

cursor = conn.cursor()                      
cursor.execute('''select CRSDepTime, avg(arrdelay) as arrdelay
from ontime
where month=10 and dayofweek=1
group by CRSDepTime
order by avg(arrdelay)
limit 1''')    
rows = cursor.fetchall()         
for row in rows:
    print(row)

cursor.close()
conn.close()
```
